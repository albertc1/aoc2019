```python

with open('input.txt') as f:
    input_string = f.read()

def parse(string):
    empty = set()
    asteroids = set()
    setmap = {
        '#': asteroids,
        '.': empty,
    }
    for y, row in enumerate(string.split()):
        for x, n in enumerate(row.strip()):
            setmap[n].add((x, y))
    return asteroids, empty
        
```

```python
asteroids, _ = parse(input_string)

best = None
max_size = 0
for x, y in asteroids:
    slopes = set()
    for x1, y1, in asteroids:
        if (x, y) == (x1, y1):
            continue
        if x1 == x:
            slope = 10000000
            direction = -1 if y1 < y else 1            
        else:
            slope = (y1 - y) / (x1 - x)
            direction = -1 if x1 < x else 1
        slopes.add((slope, direction))
    if len(slopes) > max_size:
        max_size = len(slopes)
        best = (x, y)
best, max_size
```

```python
from collections import defaultdict

amap = defaultdict(list)
x, y = best
for x1, y1, in asteroids:
    if (x, y) == (x1, y1):
        continue
    dx = x1 - x
    dy = y - y1

    if dx == 0:
        slope = 10000000
        direction = -1 if y < y1 else 1            
    else:
        slope = dy / dx
        direction = -1 if x1 < x else 1
    amap[(direction, slope)].append((dx*dx + dy*dy, x1, y1))
```

```python
for a in amap.values():
    a.sort(reverse=True)

buckets = sorted(amap.keys(), reverse=True)
```

```python
done = min(200, len(asteroids) - 1)
processed = 0
while processed < done:
    for b in buckets:
        if not amap[b]:
            continue
        d, x1, y1 = amap[b].pop()
        processed += 1
        print(processed, b, d, x1, y1)

            
    
```
